```python
def word_ratings(app_rating):

    word_rating = ''

    rating_dict = {5.0 : 'OutStandingApp', 4.5 : 'ExcellentApp', 4.0 : 'GreatApp', 3.5 : 'GoodApp', 3.0 : 'OkApp', 
                   2.5 : 'MediocreApp', 2.0 : 'PoorApp', 1.5 : 'BadApp', 1.0 : 'HorribleApp', 0.0 : 'AbysmalApp'}

    if app_rating in rating_dict:
        word_rating = rating_dict[app_rating] 

    return word_rating
```

```python
def word_total_ratings(total_ratings):
    
    total_word_rating = ''

    if total_ratings >= 0 and total_ratings <= 999:
        total_word_rating = 'Range1'
        return total_word_rating

    if total_ratings >= 1000 and total_ratings <= 9999:
        total_word_rating = 'Range2'
        return total_word_rating

    if total_ratings >= 10000 and total_ratings <= 24999:
        total_word_rating = 'Range3'
        return total_word_rating

    if total_ratings >= 25000 and total_ratings <= 49999:
        total_word_rating = 'Range4'
        return total_word_rating

    if total_ratings >= 25000 and total_ratings <= 49999:
        total_word_rating = 'Range4'
        return total_word_rating

    if total_ratings >= 50000 and total_ratings <= 74999:
        total_word_rating = 'Range5'
        return total_word_rating


    if total_ratings >= 75000 and total_ratings <= 99999:
        total_word_rating = 'Range6'
        return total_word_rating

    if total_ratings >= 100000 and total_ratings <= 499999:
        total_word_rating = 'Range7'
        return total_word_rating

    if total_ratings >= 500000 and total_ratings <= 999999:
        total_word_rating = 'Range8'
        return total_word_rating

    if total_ratings >= 1000000 and total_ratings <= 9999999:
        total_word_rating = 'Range9'
        return total_word_rating

    if total_ratings >= 10000000:
        total_word_rating = 'Range10'
        return total_word_rating



    # 0 - 999 - Range1
    # 1000 - 9999 - Range2
    # 10000 - 24999 - Range3
    # 25000 - 49999 - Range4
    # 50000 - 74999 - Range5
    # 75000 - 99999 - Range6
    # 100000 - 499999 - Range7
    # 500000 - 999999 - Range8
    # 1M+ - Range9
    # 10M+ - Range10
```

```python
def word_price(price):

    word_price = ''

    if price == 0:
        word_price = 'FreeApp'
        return word_price

    if price > 0 and price <= 9.99:
        word_price = 'PaidAppRange1'
        return word_price

    if price >= 10 and price <= 99.99:
        word_price = 'PaidAppRange2'
        return word_price

    if price >= 100:
        word_price = 'PaidAppRange3'
        return word_price
```

```python
def text_processing(app_description):
    
    # Remove Bullet Point, Star & '—'(Can remove any of them like these if found)
    def remove_values_from_list(the_list, val) : return [value for value in the_list if value != val]

    if '•' in app_description.split():
        app_description_split = app_description.split()
        app_description_split = remove_values_from_list(app_description_split, '•')
        app_description = ' '.join(app_description_split)

    if '★' in app_description.split():
        app_description_split = app_description.split()
        app_description_split = remove_values_from_list(app_description_split, '★')
        app_description = ' '.join(app_description_split)

    if '—' in app_description.split():
        app_description_split = app_description.split()
        app_description_split = remove_values_from_list(app_description_split, '—')
        app_description = ' '.join(app_description_split)
    
    # Remove New Line Character
    app_description = ''.join(app_description.splitlines())

    # Remove Punctuation
    app_description = [char for char in app_description if char not in string.punctuation]
    app_description = ''.join(app_description)
    
    return app_description
```

```python
def list_stem_and_string(tags):

    word = []

    for value in tags:
        word.append(ps.stem(value))

    return " ".join(word)
```

```python
def apps_recommend(app_name):
    app_index = new[new['app_name'] == app_name].index[0]
    app_distances = sorted(list(enumerate(apps_similarity[app_index])),reverse = True,key = lambda x : x[1])
    for i in app_distances[1:6]:
        print(new.iloc[i[0]].app_name)
```
